```python
def resize(sample):
    nrow, ncol,_ = sample.shape
    mindim = min(nrow, ncol)
    resized = sample[(nrow - mindim) // 2 : (nrow + mindim) // 2,
                    (ncol - mindim) //2 : (ncol + mindim) // 2,:]
    return resized
```

```python
def rot(mat, phi):
    rot = np.array([
        [np.cos(phi), -1*np.sin(phi)],
        [np.sin(phi),    np.cos(phi)]
    ])
    return signal.convolve2d(mat, rot)
```
